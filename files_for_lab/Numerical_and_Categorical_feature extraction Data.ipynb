## Remove outliers

```python
# function to remove outliers
def remove_outliers(numerical, threshold=1.5, in_columns = ['customer_lifetime_value', 'monthly_premium_auto', 'total_claim_amount']):
    for column in in_columns:
        upper = np.percentile(numerical[column], 75)
        lower = np.percentile(numerical[column], 25)
        iqr = upper - lower
        upper_limit = upper + (threshold * iqr)
        lower_limit = lower - (threshold * iqr)
        numerical = numerical[(numerical[column] > lower_limit) & (numerical[column] < upper_limit)]
    return customer_df

numerical.head()
```

## Check for redundant values

```python
for i in categoricals.columns:
    print(categoricals[i].value_counts())
    print()
    print('--------------')
```

```python
def clean_employment_status(df):
    df = df.replace('Medical Leave', 'Employed')
    return df

categoricals['employment_status'] = categoricals['employment_status'].apply(clean_employment_status)
categoricals['employment_status'].value_counts()
```

```python
def clean_policy_type(df):
    df = df.replace(' ', '')
    df = df.replace('Auto', '')
    return df

categoricals['policy_type'] = categoricals['policy_type'].apply(clean_policy_type)
categoricals['policy_type'].value_counts()
```

```python
def clean_policy(df):
    dict = {' ': '', 'Corporate': '', 'Personal': '', 'Special':''}
    for key in dict.keys():
        df = df.replace(key, dict[key])
    return df

categoricals['policy'] = categoricals['policy'].apply(clean_policy)
categoricals['policy'].value_counts()
```

```python
def clean_vehicle_class(df):
    df = df.replace('Luxury SUV', 'Luxury Car')
    return df

categoricals['vehicle_class'] = categoricals['vehicle_class'].apply(clean_vehicle_class)
categoricals['vehicle_class'].value_counts()
```
