```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
pd.set_option('display.max_columns', None)
```

## Load the dataset

```python
customer_df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
customer_df.head()
```

```python
customer_df.info()
```

```python
customer_df.shape
```

## Rename the columns 

```python
customer_df.columns = [col.lower() for col in customer_df]
```

```python
customer_df.columns = [c.replace(' ', '_') for c in customer_df]
```

```python
customer_df = customer_df.rename(columns={'employmentstatus':'employment_status'})
```

```python
customer_df.head()
```

## Fix the data types 

```python
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'], errors='coerce')
```

```python
for col in ['customer', 'state', 'response', 'coverage', 'education', 'employment_status', 'gender', 'location_code', 'marital_status',
           'policy_type', 'policy', 'renew_offer_type', 'sales_channel', 'vehicle_class', 'vehicle_size']:
    customer_df[col] = customer_df[col].astype('category')
```

```python
customer_df.dtypes
```

## Plot a correlation matrix

```python
corr_matrix = customer_df.corr()
corr_matrix
```

```python
plt.figure(figsize=(18, 8))
sns_plot = sns.heatmap(corr_matrix, annot=True)
figure = sns_plot.get_figure()
plt.show()
```

## Plot every continuous variable.

```python
numerical = customer_df._get_numeric_data()
numerical.head()
```

```python
for column in numerical.columns:
    sns.distplot(numerical[column], bins=20)
    plt.title(column)
    plt.show()
```

## categorical variables

```python
categoricals = customer_df.select_dtypes('category')
categoricals.head()
```

```python
categoricals = categoricals.drop(labels='customer', axis=1)
categoricals.head()
```

```python
for i in categoricals.columns:
    plt.figure(figsize=(18, 8))
    sns.countplot(x=i, data=categoricals)
    plt.title('Distribution of {}'.format(i))
    plt.xticks(rotation=45)
    plt.show()
```

## Look for outliers 

```python
for i in numerical.columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=i, data=numerical)
    plt.title('Boxplot for {}'.format(i))
    plt.show()
```
